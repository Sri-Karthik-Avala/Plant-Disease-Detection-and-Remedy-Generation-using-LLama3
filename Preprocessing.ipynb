```python


import os
from PyPDF2 import PdfReader, PdfWriter

# Define the paths
input_folder = r'C:\Users\srika\OneDrive\Desktop\New folder\planty\temp_pdfs'
output_pdf = r'C:\Users\srika\OneDrive\Desktop\New folder\planty\final_merged.pdf'

# Create a PDF writer object
pdf_writer = PdfWriter()

# Get a list of all PDF files in the input folder
pdf_files = [f for f in os.listdir(input_folder) if f.endswith('.pdf')]

# Sort the files if needed
pdf_files.sort()

# Merge all PDF files
for pdf_file in pdf_files:
    pdf_path = os.path.join(input_folder, pdf_file)
    try:
        pdf_reader = PdfReader(pdf_path)
        for page in range(len(pdf_reader.pages)):
            pdf_writer.add_page(pdf_reader.pages[page])
        print(f"Added {pdf_file} to the final PDF.")
    except Exception as e:
        print(f"Failed to add {pdf_file}: {e}")

# Write out the merged PDF
with open(output_pdf, 'wb') as output_pdf_file:
    pdf_writer.write(output_pdf_file)

print(f"Merged PDF saved as {output_pdf}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, GlobalAveragePooling2D
from tensorflow.keras.utils import plot_model

# Define the number of classes
num_classes = 88  # Adjust this to the number of classes in your problem

# Load the DenseNet201 model with pre-trained ImageNet weights
base_model = tf.keras.applications.DenseNet201(input_shape=(224, 224, 3),
                                               include_top=False,
                                               weights='imagenet')

# Create the model
model = Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Flatten())
model.add(Dense(1024, activation="relu"))
model.add(Dense(512, activation="relu"))
model.add(Dense(num_classes, activation="softmax", name="classification"))

# Print the model summary
model.summary()

# Generate and save the model diagram
plot_model(model, to_file='model_diagram.png', show_shapes=True, show_layer_names=True)
```
